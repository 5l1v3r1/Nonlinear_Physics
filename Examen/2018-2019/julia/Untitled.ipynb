```julia
using LinearAlgebra
using Plots; pyplot()
```

# Exercice 1 : La courbe de Koch

```julia
function Koch_curve()
    
    # --> Define the various linear transformations.    
    A₁ = [1/3 0.0 ; 0.0 1/3]
    b₁ = [0.0 ; 0.0]
    
    function f₁(x)
        A₁ = [1/3 0.0 ; 0.0 1/3]
        b₁ = [0.0 ; 0.0]
        return A₁*x .+ b₁
    end
    
    function f₂(x)
        A₂ = [1/6 -√3/6 ; √3/6 1/6]
        b₂ = [1/3 ; 0]
        return A₂*x .+ b₂
    end
    
    function f₃(x)
        A₃ = [1/6 √3/6 ; -√3/6 1/6]
        b₃ = [1/2 ; √3/6]
        return A₃*x .+ b₃
    end
    
    function f₄(x)
        A₄ = [1/3 0 ; 0 1/3]
        b₄ = [2/3 ; 0]
        return A₄*x .+ b₄
    end
    
    # --> Segment initial.
    x, y = [0 ; 0] , [1 ; 0]
    
    p = plot(
            [x[1], y[1]], [x[2], y[2]],
            linewidth = 2,
            color = 1,
            )
    
    # --> First level.
    x₁, y₁ = f₁(x), f₁(y)
    
    p = plot!(
            [x₁[1], y₁[1]] , [x₁[2], y₁[2]].-1/3,
            linewidth = 2,
            color = 1,
    )

    x₂, y₂ = f₂(x), f₂(y)
    
    p = plot!(
            [x₂[1], y₂[1]] , [x₂[2], y₂[2]].-1/3,
            linewidth = 2,
            color = 2,        
    )

    x₃, y₃ = f₃(x), f₃(y)
    
    p = plot!(
            [x₃[1], y₃[1]] , [x₃[2], y₃[2]].-1/3,
            linewidth = 2,
            color = 3,        
    )
    
    x₄, y₄ = f₄(x), f₄(y)
    
    p = plot!(
            [x₄[1], y₄[1]] , [x₄[2], y₄[2]].-1/3,
            linewidth = 2,
            color = 4,  
            aspect_ratio = :equal,
            legend = :none,
    )

    # --> Second level.
    X = hcat(x₁ , x₂, x₃, x₄)
    Y = hcat(y₁ , y₂ , y₃ , y₄)
    
    X₁, Y₁ = f₁(X), f₁(Y)
    
    p = plot!(
        [X₁[1, :], Y₁[1, :]] , [X₁[2, :].-2/3, Y₁[2, :].-2/3],
        linewidth = 2,
        color = 1,
    )
        
    X₂, Y₂ = f₂(X), f₂(Y)
    
    p = plot!(
        [X₂[1, :], Y₂[1, :]] , [X₂[2, :].-2/3, Y₂[2, :].-2/3],
        linewidth = 2,
        color = 2,
    )
        
    X₃, Y₃ = f₃(X), f₃(Y)
    
    p = plot!(
        [X₃[1, :], Y₃[1, :]] , [X₃[2, :].-2/3, Y₃[2, :].-2/3],
        linewidth = 2,
        color = 3,
    )
        
    X₄, Y₄ = f₄(X), f₄(Y)
    
    p = plot!(
        [X₄[1, :], Y₄[1, :]] , [X₄[2, :].-2/3, Y₄[2, :].-2/3],
        linewidth = 2,
        color = 4,
        yaxis = false,
        grid = false,
    )
    
    # --> Third level.
    X = hcat(X₁, X₂, X₃, X₄)
    Y = hcat(Y₁, Y₂, Y₃, Y₄)

        
    X₁, Y₁ = f₁(X), f₁(Y)
    
    p = plot!(
        [X₁[1, :], Y₁[1, :]] , [X₁[2, :].-3/3, Y₁[2, :].-3/3],
        linewidth = 2,
        color = 1,
    )
        
    X₂, Y₂ = f₂(X), f₂(Y)
    
    p = plot!(
        [X₂[1, :], Y₂[1, :]] , [X₂[2, :].-3/3, Y₂[2, :].-3/3],
        linewidth = 2,
        color = 2,
    )
        
    X₃, Y₃ = f₃(X), f₃(Y)
    
    p = plot!(
        [X₃[1, :], Y₃[1, :]] , [X₃[2, :].-3/3, Y₃[2, :].-3/3],
        linewidth = 2,
        color = 3,
    )
        
    X₄, Y₄ = f₄(X), f₄(Y)
    
    p = plot!(
        [X₄[1, :], Y₄[1, :]] , [X₄[2, :].-3/3, Y₄[2, :].-3/3],
        linewidth = 2,
        color = 4,
        yaxis = false,
        grid = false,
        xlabel = "x", xticks = 0:0.5:1,
        size = (800, 800),
        tickfont = 20, guidefont = 20,
    )
    
    # --> Fourth level.
    X = hcat(X₁, X₂, X₃, X₄)
    Y = hcat(Y₁, Y₂, Y₃, Y₄)

        
    X₁, Y₁ = f₁(X), f₁(Y)
    
    p = plot!(
        [X₁[1, :], Y₁[1, :]] , [X₁[2, :].-4/3, Y₁[2, :].-4/3],
        linewidth = 2,
        color = 1,
    )
        
    X₂, Y₂ = f₂(X), f₂(Y)
    
    p = plot!(
        [X₂[1, :], Y₂[1, :]] , [X₂[2, :].-4/3, Y₂[2, :].-4/3],
        linewidth = 2,
        color = 2,
    )
        
    X₃, Y₃ = f₃(X), f₃(Y)
    
    p = plot!(
        [X₃[1, :], Y₃[1, :]] , [X₃[2, :].-4/3, Y₃[2, :].-4/3],
        linewidth = 2,
        color = 3,
    )
        
    X₄, Y₄ = f₄(X), f₄(Y)
    
    p = plot!(
        [X₄[1, :], Y₄[1, :]] , [X₄[2, :].-4/3, Y₄[2, :].-4/3],
        linewidth = 2,
        color = 4,
        yaxis = false,
        grid = false,
        xlabel = "x", xticks = 0:0.5:1,
        size = (800, 1200),
        tickfont = 20, guidefont = 20,
    )
    
    return p
end
```
